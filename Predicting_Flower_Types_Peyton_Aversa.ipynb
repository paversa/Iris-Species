```python
import warnings
warnings=warnings.filterwarnings('ignore')    #ignore warnings
import pandas as pd
import matplotlib.pyplot as plt      #import pandas, matplotlin, and seaborn
%matplotlib inline
import seaborn as sns
sns.set(style="white", color_codes=True)
```

```python
path="/content/drive/MyDrive/Nittany AI/Colab Notebooks/Project 1/Iris.csv"
iris=pd.read_csv(path)
iris.head(5)
```

```python
iris["Species"].value_counts()
```

```python
sns.FacetGrid(iris, hue="Species",height=6).map(plt.scatter,"SepalLengthCm", "SepalWidthCm").add_legend()
```

Logistic Regression

```python
flower_mapping = {'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2}
iris["Species"]=iris["Species"].map(flower_mapping)
```

```python
iris.head()
```

Preparing Inputs and Outputs

```python
X=iris[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].values
y=iris[['Species']].values
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(X,y)
```

Accuracy

```python
model.score(X,y)
```

```python
expected=y
predicted=model.predict(X)
predicted
```

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(expected,predicted))
```

```python
print(metrics.confusion_matrix(expected,predicted))
```

Regularization

```python
model=LogisticRegression(C=20,penalty='12')
model.fit(X,y)
model.score(X,y)
```
